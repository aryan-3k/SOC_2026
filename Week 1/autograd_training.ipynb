```python
import torch
import matplotlib.pyplot as plt
```

```python

# ==========================================
# 1. What is a Gradient?
# ==========================================

# A simple example: y = x^2, dy/dx = 2x
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
y.backward()   # compute gradients
print("x      =", x.item())
print("y      =", y.item())
print("dy/dx  =", x.grad.item())   # should be 2*3 = 6
print("-" * 40)

# TODO: Compute the gradient of y = x^3 at x = 2.0
#       The answer should be 3 * 2^2 = 12
x_custom = torch.tensor(2.0, requires_grad=True)
y_custom = x_custom ** 3
y_custom.backward()
print("x_custom     =", x_custom.item())
print("y_custom     =", y_custom.item())
print("dy_custom/dx =", x_custom.grad.item()) # should be 12.0
print("=" * 40)

```

```python

# ==========================================
# 2. A Simple Linear Model
# ==========================================

torch.manual_seed(42)
# True relationship we want the model to discover: y = 2x + 1
X = torch.linspace(-1, 1, 20).unsqueeze(1)   # shape (20, 1)
y = 2 * X + 1 + 0.1 * torch.randn_like(X)    # add a tiny bit of noise

# Initialise learnable parameters
w = torch.randn(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

lr = 0.1   # learning rate
losses = []
for epoch in range(100):
    # Forward pass
    y_pred = w * X + b

    # Loss: Mean Squared Error
    loss = ((y_pred - y) ** 2).mean()
    losses.append(loss.item())

    # Backward pass — compute gradients
    loss.backward()

    # Update weights (gradient descent step)
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad

    # Zero gradients before next iteration — IMPORTANT
    w.grad.zero_()
    b.grad.zero_()

print(f"Learned w = {w.item():.3f}  (true = 2.0)")
print(f"Learned b = {b.item():.3f}  (true = 1.0)")

# TODO: Change the learning rate to 0.01 and rerun. What happens to the loss curve?
#       Then try 0.5. What happens? Write your observations as a comment below.

# OBSERVATIONS:
# 1. With lr = 0.01: The loss curve drops much more slowly. 100 epochs are not enough 
#    for the parameters to fully converge to their true values (w and b stay further from 2.0 and 1.0).
# 2. With lr = 0.5: The loss curve drops incredibly fast in the first few epochs. 
#    However, if it's set too high, the loss can oscillate or even diverge because 
#    the updates overshoot the minimum.
print("=" * 40)

```

```python

# ==========================================
# 3. Using torch.optim
# ==========================================

w = torch.randn(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = torch.optim.Adam([w, b], lr=0.1)
losses = []
for epoch in range(100):
    y_pred = w * X + b
    loss = ((y_pred - y) ** 2).mean()
    losses.append(loss.item())

    optimizer.zero_grad()   # zero gradients
    loss.backward()         # compute gradients
    optimizer.step()        # update weights

print("Adam Optimization Results:")
print(f"Learned w = {w.item():.3f}  (true = 2.0)")
print(f"Learned b = {b.item():.3f}  (true = 1.0)")
print("-" * 40)

# TODO: Replace Adam with torch.optim.SGD and compare the final w and b.
#       Which converges faster?
w_sgd = torch.randn(1, requires_grad=True)
b_sgd = torch.zeros(1, requires_grad=True)

optimizer_sgd = torch.optim.SGD([w_sgd, b_sgd], lr=0.1)
losses_sgd = []

for epoch in range(100):
    y_pred_sgd = w_sgd * X + b_sgd
    loss_sgd = ((y_pred_sgd - y) ** 2).mean()
    losses_sgd.append(loss_sgd.item())

    optimizer_sgd.zero_grad()
    loss_sgd.backward()
    optimizer_sgd.step()

print("SGD Optimization Results:")
print(f"Learned w (SGD) = {w_sgd.item():.3f}  (true = 2.0)")
print(f"Learned b (SGD) = {b_sgd.item():.3f}  (true = 1.0)")

# COMPARISON OBSERVATION:
# Adam generally converges faster than vanilla SGD at the same learning rate 
# because Adam adapts the learning rate dynamically for every parameter based on 
# running moments of the gradients, whereas SGD uses a fixed step size.
print("=" * 40)

```

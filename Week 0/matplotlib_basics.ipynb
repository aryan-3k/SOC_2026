```python
### Assignment 1: Double Histogram

'''
We require you to generate and compare two datasets. First, using the function numpy.random.normal(), from a 
normal distribution with mean 0 and std = 1, create an array of length 1000. Next, using the function numpy.random.beta(),
from a beta distribution with a=2, b=5 create an array of length 1000. Finally, plot both arrays as histograms
(20 bins) side by side ensuring correct titles and axis labels using Matplotlib subplots. Brownie points if
you customize the plots so that the final figure looks aesthetic.
'''
def assignment1():
    x = random.normal(loc=0.00, scale=1.00, size=1000)
    y = random.beta(a=2, b=5, size=1000)
    plt.subplot(1,2,2)
    plt.hist(y,bins=20, color='green', edgecolor='black')
    plt.title("Beta Distribution")
    plt.xlabel("Probability / Proportion")
    plt.subplot(1,2,1)
    plt.title("Normal Distribution")
    plt.hist(x, bins=20, color='orange', edgecolor='black')
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.suptitle("Datasets")
    plt.tight_layout()
    plt.show()
assignment1()
```

```python
### Assignment 2: Histogram with lines

'''
Using the function numpy.random.exponential(). Use it to create a NumPy array of length 500 named 'data_exp'
with a scale parameter of two. Plot a histogram of this data, adding vertical lines to clearly indicate the
median and the 90th percentile of the data_exp array. Ensure the plot has a clear title, axis labels,
and a legend identifying the line markers. Again brownie points if you manage to customize this plot to look good.
'''
def assignment2():
    data_exp = random.exponential(scale=2, size=500)
    plt.hist(data_exp, bins=30, color='green', edgecolor='black')
    plt.axvline(np.median(data_exp), color='red', linestyle='--', lw=2, label=f'Median: {np.median(data_exp):.2f}')
    plt.axvline(np.percentile(data_exp,90), color='orange', linestyle=':', lw=2, label=f'90th Percentile: {np.percentile(data_exp,90):.2f}')
    plt.grid(axis='y')
    plt.title("Exponential Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()  
assignment2    
```
